# Task
Construct the Lagrange basis polynomials and the interpolated polynomial for the function sin(x) at points x = -π/4, 0, π/2. Evaluate the interpolated polynomial at x = π/3. Use the `Polynomial` class from the NumPy library.

## Define the data points

### Subtask:
Define the x-values and corresponding y-values (sin(x)) for the given points.


**Reasoning**:
Define the x-values and their corresponding sine values as NumPy arrays.



```python
x_points = np.array([-np.pi/4, 0, np.pi/2])
y_points = np.sin(x_points)
```

## Construct lagrange basis polynomials

### Subtask:
For each data point, construct the corresponding Lagrange basis polynomial using the `Polynomial` class.


**Reasoning**:
Construct the Lagrange basis polynomials for each data point using the Polynomial class and store them in a list.



```python
lagrange_basis_polynomials = []
for k in range(len(x_points)):
    l_k_numerator = Polynomial([1])  # Initialize as 1
    l_k_denominator = 1

    for j in range(len(x_points)):
        if k != j:
            # Numerator term (x - x_j)
            term_numerator = Polynomial([-x_points[j], 1])
            l_k_numerator = l_k_numerator * term_numerator

            # Denominator term (x_k - x_j)
            l_k_denominator *= (x_points[k] - x_points[j])

    l_k = l_k_numerator / l_k_denominator
    lagrange_basis_polynomials.append(l_k)

for i, poly in enumerate(lagrange_basis_polynomials):
    print(f"l_{i}(x) = {poly}")
```

## Construct the interpolated polynomial

### Subtask:
Combine the basis polynomials and the y-values to form the interpolated polynomial.


**Reasoning**:
Combine the basis polynomials and the y-values to form the interpolated polynomial.



```python
interpolated_polynomial = Polynomial([0])
for i in range(len(y_points)):
    interpolated_polynomial += y_points[i] * lagrange_basis_polynomials[i]

print(interpolated_polynomial)
```

## Evaluate the interpolated polynomial

### Subtask:
Evaluate the interpolated polynomial at the specified point.


**Reasoning**:
Evaluate the interpolated polynomial at x = π/3 and print the result.



```python
x_eval = np.pi / 3
interpolated_value = interpolated_polynomial(x_eval)
print(interpolated_value)
```

## Calculate central difference from forward and backward differences

### Subtask:
Use the given forward and backward difference approximations to compute the central difference approximation.


**Reasoning**:
Calculate the central difference approximation using the given forward and backward difference approximations and print the result.



```python
forward_difference = 2.4
backward_difference = 2.3
central_difference = (forward_difference + backward_difference) / 2
print(central_difference)
```

## Implement central difference function

### Subtask:
Write a Python function that takes a NumPy array of x-values and a step size h, and returns a NumPy array of the central difference approximations of the derivative of $\sin(x)$.


**Reasoning**:
Define the `central_difference_sin` function to compute the central difference approximation of the derivative of sin(x) for a given NumPy array and step size.



```python
def central_difference_sin(x_values, h):
  """
  Computes the central difference approximation of the derivative of sin(x).

  Args:
    x_values: A NumPy array of x-values.
    h: The step size.

  Returns:
    A NumPy array of the central difference approximations.
  """
  return (np.sin(x_values + h) - np.sin(x_values - h)) / (2 * h)
```

## Compute and print central difference for specific points

### Subtask:
Use the implemented function to compute and print the central difference approximation for the given x-values.


**Reasoning**:
Define the x-values and step size and then compute the central difference approximation using the defined function.



```python
x_specific = np.array([0, np.pi/3, np.pi/2])
h = 0.5
central_diff_approx = central_difference_sin(x_specific, h)
print(central_diff_approx)
```

Calculate the central difference approximation and the exact derivative for the generated x-values.



```python
central_diff_plot = central_difference_sin(x_plot, h=0.01)
exact_derivative_plot = np.cos(x_plot)
```

## Plot the approximations
